```python
import numpy as np
import open3d as o3d
import os

# 1. 加载 npy 深度文件
depth_path = '/home/tianqi/ws_realsense/captured_images/depth_20251221_015605.npy'
rgb_path = '/home/tianqi/ws_realsense/captured_images/rgb_20251221_015605.jpg'
depth_data = np.load(depth_path)

# 2. 配置相机内参 (fx, fy, cx, cy)
intrinsic = o3d.camera.PinholeCameraIntrinsic(
    width=848, height=480, 
    fx=422.4, fy=422.4, cx=424.0, cy=240.0
)

# 3. 将深度图转换为点云
# depth_scale=1000 表示深度值单位是毫米，转换为米
depth_o3d = o3d.geometry.Image(depth_data.astype(np.float32))
pcd = o3d.geometry.PointCloud.create_from_depth_image(
    depth_o3d, intrinsic, depth_scale=1000.0, depth_trunc=3.0
)

# 4. 可视化点云
print(f"成功读取 {depth_path}，正在可视化...")
o3d.visualization.draw_geometries([pcd])
```
